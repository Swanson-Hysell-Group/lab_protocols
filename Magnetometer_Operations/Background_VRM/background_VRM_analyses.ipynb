# This notebook is for analyzing the backgournd drifting of the three axes SQUIDs on the magnetometer

- 05/27/2021 overnight VRM acquisition recorded in file "background_measurement_05272021.csv"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

## 2021 May background test

```python
bacckground_05272021 = pd.read_csv('background_measurement_05272021.csv', header=0)
bacckground_05272021.head()
```

```python
plt.figure(figsize=(18, 8))
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['x'], label='x', color = 'C0')
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['y'], label='y', color = 'C1')
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['z'], label='z', color = 'C2')

plt.title('time series background drfting on all three axes', fontsize=18)
plt.xlabel('time (s)', fontsize=12)
plt.ylabel('background reading', fontsize=12)
plt.legend()
plt.savefig('all_axes_drift.pdf')
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['x'], label='x', color = 'C0')

plt.title('time series background drfting on x axis', fontsize=18)
plt.xlabel('time (s)', fontsize=12)
plt.ylabel('background reading', fontsize=12)
plt.legend()
plt.savefig('x_axis_drift.pdf')
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['y'], label='y', color = 'C1')

plt.title('time series background drfting on y axis', fontsize=18)
plt.xlabel('time (s)', fontsize=12)
plt.ylabel('background reading', fontsize=12)
plt.legend()
plt.savefig('y_axis_drift.pdf')
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.scatter(bacckground_05272021['Time (s)'], bacckground_05272021['z'], label='z', color = 'C2')

plt.title('time series background drfting on z axis', fontsize=18)
plt.xlabel('time (s)', fontsize=12)
plt.ylabel('background reading', fontsize=12)
plt.legend()
plt.savefig('z_axis_drift.pdf')
plt.show()
```
